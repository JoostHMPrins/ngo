```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.load('../../../trainingdata/HC2D_NGO_VarMiONpapersettings/x.npy')
u = np.load('../../../trainingdata/HC2D_NGO_VarMiONpapersettings/u.npy')

x_rand = np.load('../../../trainingdata/HC2D_NGO_randoml_negativef/x.npy')
u_rand = np.load('../../../trainingdata/HC2D_NGO_randoml_negativef/u.npy')
```

```python
U, Sigma, VT = np.linalg.svd(u, full_matrices=True)
```

```python
U, Sigma_rand, VT = np.linalg.svd(u_rand, full_matrices=True)
```

```python
size_title = 25
size_ticklabels = 15
size_axeslabels = 20

fig, ax = plt.subplots(1,1, figsize=(6, 4))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.4, hspace=0.4)

ax.plot(np.linspace(1,len(Sigma),len(Sigma)),Sigma/Sigma[0], linewidth=3, label="VarMiON dataset")
ax.plot(np.linspace(1,len(Sigma_rand),len(Sigma_rand)),Sigma_rand/Sigma_rand[0], linewidth=3, label="New dataset")
ax.legend(fontsize=12)
ax.tick_params(axis='both', labelsize=size_ticklabels)
ax.ticklabel_format(style='sci', scilimits=[-3,3])
ax.set_xlabel(r'$i$', fontsize=size_axeslabels)
ax.set_ylabel(r'$\lambda_i/\lambda_1$', fontsize=size_axeslabels)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-6,1)
ax.set_xlim(1,1e3)
ax.grid()

plt.savefig("eigenvaluedecay.svg", bbox_inches='tight')#, transparent=True)
```

```python
size_title = 25
size_ticklabels = 15
size_axeslabels = 20

fig, ax = plt.subplots(1,1, figsize=(6, 4))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.4, hspace=0.4)

ax.plot(np.linspace(1,len(Sigma),len(Sigma)),np.cumsum(Sigma**2)/np.sum(Sigma**2), linewidth=3, label="VarMiON dataset")
ax.plot(np.linspace(1,len(Sigma_rand),len(Sigma_rand)),np.cumsum(Sigma_rand**2)/np.sum(Sigma_rand**2), linewidth=3, label="New dataset")
ax.legend(fontsize=12)
ax.tick_params(axis='both', labelsize=size_ticklabels)
ax.ticklabel_format(style='sci', scilimits=[-3,3])
ax.set_xlabel(r'$i$', fontsize=size_axeslabels)
ax.set_ylabel(r'$\sum_{s=1}^i\lambda^2_s/\sum_{s=1}^n\lambda^2_s$', fontsize=size_axeslabels)
ax.set_xscale('log')
ax.set_yscale('log')
# ax.set_xlim(1,1e3)
ax.grid()

plt.savefig("eigenvaluecumsum.svg", bbox_inches='tight')#, transparent=True)
```
