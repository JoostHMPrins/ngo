```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
F = torch.tensor(np.load('F.npy'))
```

```python
i = 0
plt.imshow(F[i])
plt.colorbar()
```

```python
lambdas = torch.linalg.eigvals(F)
lambdas_abs = torch.real((lambdas*torch.conj(lambdas))**(1/2))
lambda_max = torch.amax(lambdas_abs, axis=-1)
scaling = lambda_max
F_scaled = F/scaling[:,None,None]
```

```python
i = 4
plt.imshow(F[i])
plt.colorbar()
plt.show()

plt.imshow(F_scaled[i])
plt.colorbar()
plt.show()
```
