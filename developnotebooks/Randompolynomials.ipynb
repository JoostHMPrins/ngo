```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 100
x,y = np.meshgrid(np.linspace(0,1,size),np.linspace(0,1,size),indexing='xy')
```

```python
c = np.random.uniform(-1,1,10)
c
f = c[0] + c[1]*x + c[2]*y + c[3]*x**2 + c[4]*y**2 + c[5]*x*y + c[6]*x**3 + c[7]*y**3 + c[8]*x*y**2 + c[9]*y*x**2
```

```python
plt.contourf(x, y, f)
plt.colorbar()
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(x, y, f)
```

```python
def randompoly1DO3(x):
    c = np.random.uniform(-1,1,4)
    poly = c[0] + c[1]*x + c[2]*x**2 + c[3]*x**3
    return poly
```
