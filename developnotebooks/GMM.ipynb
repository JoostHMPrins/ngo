```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 0
l = 0.05
d = 2
N = int(1/l**d)
rho_n = np.random.uniform(0,1,size=N)
rho_n = rho_n/np.sum(rho_n)
mu_n = np.random.uniform(0,1,size=(N,d))
def GMM(x):
    GMM_n = rho_n[:,None]/(np.sqrt(2*np.pi)*l)*np.exp(-1/2*np.sum((x[None,:,:] - mu_n[:,None,:])**2, axis=-1)/l**2)
    # GMM_n = (np.sqrt(2*np.pi)*l)*np.exp(-1/2*np.sum((x[None,:,:] - mu_n[:,None,:])**2, axis=-1)/l**2)
    # GMM_n = rho_n[:,None]*np.exp(-1/2*np.sum((x[None,:,:] - mu_n[:,None,:])**2, axis=-1)/l**2)    
    GMM = np.sum(GMM_n, axis=0)
    return GMM
```

```python
X, Y = np.mgrid[0:1:100*1j, 0:1:100*1j]
x = np.vstack([X.ravel(), Y.ravel()]).T
```

```python
plt.tripcolor(x[:,0], x[:,1], GMM(x))
plt.axis('equal')
plt.colorbar()

# plt.savefig("GMM.svg", bbox_inches='tight', transparent=True)
```
